```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score,classification_report,confusion_matrix)
```

```python
# Load datasets
train_data = pd.read_csv('/content/fraudTrain.csv')
test_data = pd.read_csv('/content/fraudTest.csv')
```

```python
# Display datasets
print("Train Dataset")
train_data.head()
```

```python
print("Test Dataset")
test_data.head()
```

```python
# Check shapes
print("Train Shape :", train_data.shape)
print("Test Shape :", test_data.shape)
```

```python
# Check missing values
print(train_data.isnull().sum())
```

```python
print(test_data.isnull().sum())
```

```python
# Check fraud distribution
print(train_data['is_fraud'].value_counts())
```

```python
# Plot graph
train_data['is_fraud'].value_counts().plot(kind='bar',figsize=(6,4))
plt.title('Fraud vs Genuine Transactions')
plt.xlabel('is_fraud')
plt.ylabel('Count')
plt.show()
```

```python
# Remove unnecessary columns
columns_to_drop = ['Unnamed: 0','trans_date_trans_time','cc_num','first','last',
                   'street','city','state','zip','dob','trans_num']
train_data = train_data.drop(columns=columns_to_drop,errors='ignore')
test_data = test_data.drop(columns=columns_to_drop,errors='ignore')
```

```python
# Check remaining columns
print(train_data.columns)
```

```python
label_encoder = LabelEncoder()
categorical_columns = ['merchant','category','gender','job']
for col in categorical_columns:
    # Combine unique values from train and test for fitting
    combined_data = pd.concat([train_data[col], test_data[col]], axis=0).astype(str)
    label_encoder.fit(combined_data.unique())
    # Transform both train and test data
    train_data[col] = label_encoder.transform(train_data[col].astype(str))
    test_data[col] = label_encoder.transform(test_data[col].astype(str))
```

```python
#Create features and target
X_train = train_data.drop('is_fraud',axis=1)
y_train = train_data['is_fraud']
X_test = test_data.drop('is_fraud',axis=1)
y_test = test_data['is_fraud']
```

```python
#Build model
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
#Predict
y_pred = model.predict(X_test)
```

```python
#Accuracy
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy :")
print(round(accuracy*100,2),"%")
```

```python
#Classification report
print(classification_report(y_test,y_pred, zero_division=0))
```

```python
#Confusion matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# Visualize confusion matrix
plt.figure(figsize=(5,5))
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
